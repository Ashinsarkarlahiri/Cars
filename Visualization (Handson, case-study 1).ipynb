```python
### Checking the dependence of brands(country of export) with the other variables of the cars: (Chi-square test of independence)
```

```python
# brand and cylinders:
```

```python
cars['cylinders'].unique()
```

```python
cars['cylinders']=cars['cylinders'].astype(str)
```

```python
cross_1=pd.crosstab(cars['cylinders'],cars['brand'])
```

```python
cars.info()
```

```python
cross_1
```

```python
from scipy.stats import chi2_contingency
```

```python
results=chi2_contingency(cross_1)
```

```python
results # the p-value of the test is way less than 0.05, thus we can clearly reject the null hypothesis that no of cylinders of a car and to which country it would be imported for max profit is independent. They are not independent. 
```

```python
# Elbow  scree plot for deciding the number of clusters
clusters = list(range(2,11))
ssd = []
for num_clusters in clusters:
    model_clus = KMeans(n_clusters = num_clusters, max_iter=150 )
    #model_clus.fit(cars_new_predictor)
    pred=model_clus.fit(cars_new_predictor)
    print(pred.labels_)
    ssd.append(model_clus.inertia_)
```

```python
plt.plot(clusters,ssd) 
plt.ylabel('ssd')
plt.xlabel('clusters')
plt.grid()
plt.show()
```

```python
# After 4 clusters the ssd inside the clusters is decreasing in a slow rate.
```

## CONCLUSIONS FROM STUDY:
### 1)  If a car belongs to cluster 0, the highest % of it's production (and hence profit) is happening in Japan. Then 2nd preference is in Europe and the 3rd preference is in U.S.
### 2) If a car belongs to cluster 1, the highest % of it's production (and hence profit) is happening in U.S. Then in Europe (2nd pref) but by a lot less margin. And Japan is not a preference , because in Japan cars of cluster 1 are not being produced.
### 3) If a car belongs to cluster 2, the U.S is most profitable country, followed by Japan and then Europe.
### 4) If a car belongs to cluster 4 , then the 1st country of preference is U.S. 2nd is Europe and 3rd is Japan. But there is no such sig difference between production in Europe & Japan, in both the countries cluster 4 cars are more or less equally profitable.
